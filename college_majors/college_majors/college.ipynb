## Introduction

We'll explore how using the pandas plotting functionality along with the Jupyter notebook interface allows us to explore data quickly using visualizations. 

We'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by [American Community Survey](https://www.census.gov/programs-surveys/acs/), which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their [Github repo](https://github.com/fivethirtyeight/data/tree/master/college-majors).

## Assignment

- Import pandas and matplotlib (remember "% matplotlib inline")
- Read in the data and explore the head and tail
- Drop any rows with missing values

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("recent-grads.csv")

print(data.head())
```

## Assignment

Feel free to use native panda graphing methods.

- Make a scatter plot of sample size vs. median
- Make a scatter plot of sample size vs. unemployment rate
- Make a scatter plot of full time vs. median
- Make a scatter plot of men vs. median
- Make a scatter plot of women vs. median


```python
fig, axes = plt.subplots(5, 1, figsize=(15, 30))

axes[0].scatter(data['Sample_size'], data['Median'])
axes[0].set_xlabel("Sample Size")
axes[0].set_ylabel("Median")
axes[0].set_title("Sample Size vs Median")

axes[1].scatter(data['Sample_size'], data['Unemployment_rate'])
axes[1].set_xlabel("Sample Size")
axes[1].set_ylabel("Unemployment Rate")
axes[1].set_title("Sample Size vs Unemployment Rate")

axes[2].scatter(data['Full_time_year_round'], data['Median'])
axes[2].set_xlabel("Full Time")
axes[2].set_ylabel("Median")
axes[2].set_title("Full Time vs Median")

axes[3].scatter(data['Men'], data['Median'])
axes[3].set_xlabel("Men")
axes[3].set_ylabel("Median")
axes[3].set_title("Men vs Median")

axes[4].scatter(data['Women'], data['Median'])
axes[4].set_xlabel("Women")
axes[4].set_ylabel("Median")
axes[4].set_title("Women vs Median")

plt.show()
```

## Assignment

- Create a for loop that plots the histogram of sample size, median, employed, full time, unemployment rate, men and women (remember to use the exact column titles)

```python
input_params = ['Sample_size', 'Median', 'Employed', 'Unemployment_rate', 'Men', 'Women']
num_of_params = len(input_params)

fig, axes = plt.subplots(num_of_params, 1, figsize=(15, 30))

for i in range(0, num_of_params):
    axes[i].hist(data[input_params[i]].dropna(), bins = 15)
    axes[i].set_xlabel(input_params[i])
    axes[i].set_ylabel('Occurrences')
    
plt.show()
```

## Assignment

- Make a bar plot of majors vs "ShareWomen" (you should run different slices since there are many different majors)


```python
majors_by_cat = [x for _, x in data.groupby('Major_category')]
num_of_cat = len(majors_by_cat)
fig, axes = plt.subplots(num_of_cat, 1, figsize=(10, 200))
plt.subplots_adjust(hspace=1)

for i in range(0, num_of_cat):
    data_subset = majors_by_cat[i]
    axes[i].bar(data_subset['Major'], data_subset['ShareWomen'])
    for tick in axes[i].get_xticklabels():
        tick.set_rotation(90)

plt.show()
```
